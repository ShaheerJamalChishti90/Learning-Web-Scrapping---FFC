```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://realpython.github.io/fake-jobs/"
web = requests.get(url).content
soup = BeautifulSoup(web, "lxml")
```

```python
print(soup.find("h1").text.strip())
print(soup.p.text.strip())
```

```python
block1 = soup.find("div", class_="card-content")
print(block1.h2.text.strip())
print(block1.h3.text.strip())
```

```python
half_columns = soup.find_all("div", class_="column is-half")

for column in half_columns:
    title = column.find("h2").text.strip()
    company = column.find("h3").text.strip()
    location = column.find("p").text.strip()
    date_posted = column.find("time").text.strip()
    year_posted = column.find("time").text.strip()[:4]

    print(f"Title: {title}")
    print(f"Company: {company}")
    print(f"Location: {location}")
    print(f"Date Posted: {date_posted}")
    print(f"Year Posted: {year_posted}\n")

    # if date_posted <= "2020":
    #     print(f"Title: {title}")
    #     print(f"Company: {company}")
    #     print(f"Location: {location}")
    #     print(f"Year Posted: {date_posted}\n")
    # else:
    #     print(f"There are no jobs posted after 2020.\n")
```
